# Compiling Regex

```python
import re
```

```python
compiled_re = re.compile('[ab][c-y]')
compiled_re
```

### Once we have a compile regular expression, they have several methods and attributes such as



```python
string_list = ['2h4a bcxv bcsw',
 's6fn 3r9e kjgy',
 'aezg 9ae6 uiqx',
 'dvof futv egsb',
 'ahiq 3xq9 ogqk',
 '1a9u hloy yknt',
 'zdfo gfre fgto',
 'xk29 6bk6 emiw',
 'k69a ww62 gicm',
 'q0qa 5tla ahvm']
```

#### If we need more info we can use addional attributs such as:


___
## re.match() finds the start of string only and not in between
___

```python
# Additional attributes
for index,string in enumerate(string_list, start =1):
    match_regex = compiled_re.match(string)
    if match_regex:
        start = match_regex.start()  # returns start index
        group = match_regex.group()  # returns matched item
        end = match_regex.end()  # returns end index
        span = match_regex.span()  # returns a tuple(start, end) 
        print(f'{index} : {string}'.ljust(20), "-->", start, group, end, span) # returns re.match object
    else:
        print(None) # returns None if no match
        
```

___
## re.search() finds the entire string even in between and returns first match
___

```python
# Additional attributes
for index,string in enumerate(string_list, start =1):
    match_regex = compiled_re.search(string)
    if match_regex:
        start = match_regex.start()  # returns start index
        group = match_regex.group()  # returns matched item
        end = match_regex.end()  # returns end index
        span = match_regex.span()  # returns a tuple(start, end) 
        print(f'{index} : {string}'.ljust(20), "-->", start, group, end, span) # returns re.match object
    else:
        print(None) # returns None if no match
        
```

___
## re.finditer() finds the entire string even in between and returns all matches in an iterator
___

```python
# Additional attributes
for index,string in enumerate(string_list, start =1):
    match_regex = compiled_re.finditer(string)  # returns an iterator
    if match_regex:
        for match in match_regex:
            span = match.span()  # returns a tuple(start, end) 
            start = match.start()  # returns start index
            group = match.group()  # returns matched item
            end = match.end()  # returns end index
        
            print(f'Found {group} in {string}({index}) '.ljust(20), "-->",f"at {start} ending at {end} {span}") # returns re.match object, 
    else:
        print(None) # returns None if no match
        
```
