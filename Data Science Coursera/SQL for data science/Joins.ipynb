```python
## using AND clause after outer
```

```python
%%sql
select emp_id, L_name, dep_id ,dep_name,b_date
from Employees as E
left outer Join Departments as D 
on E.dep_id = dept_id_dep
AND year(B_date) < 1980 ;
```

## using Wherer clause after outer

```python
%%sql
select emp_id, L_name, dep_id ,dep_name,b_date
from Employees as E
left outer Join Departments as D 
on E.dep_id = dept_id_dep
WHERE year(B_date) < 1980 ;
```

```python
%%sql
select e.f_name, e.l_name, d.dep_name
from employees as E
full outer join departments as D on e.dep_id = d.dept_id_dep;
```
