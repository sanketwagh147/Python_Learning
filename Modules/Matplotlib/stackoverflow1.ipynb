```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
filepath = r"Stack_version_1_with_languges_binned.csv"
df = pd.read_csv(filepath)
```

```python
df.set_index("Respondent", inplace=True)
```

```python
# df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
df["Python"].sum()
```

```python
Languages_used_in_survey = ['Assembly', 'Bash/Shell/PowerShell', 'C', 'C++', 'C#', 'Clojure', 'Dart', 'Elixir', 'Erlang', 'F#', 'Go', 'HTML/CSS', 'Java', 'JavaScript', 'Kotlin', 'Objective-C', 'PHP', 'Python', 'R', 'Ruby', 'Rust', 'Scala', 'SQL', 'Swift', 'TypeScript', 'VBA', 'WebAssembly', 'Other(s):']
```

```python
lang_count_list = []
for language in Languages_used_in_survey:
    sum_ = df[language].sum()
    lang_count_list.append(sum_)
print(lang_count_list)
a_dict ={Languages_used_in_survey[i]:lang_count_list[i] for i in range(len(lang_count_list))}  # creae a dicitionay
```

```python
df1 = pd.Series(a_dict)
```

```python
df1.index.name= "Language"
```

```python
df1.rename("Count")
```

```python
df1.sort_values(inplace=True)
```

```python
df1.plot(kind="barh", figsize=(15,11), color="grey")
plt.title("Language Usage 2019 Stack Overflow Developer Survey Language Usage 2019 ")
plt.xlabel("Number of user")
plt.show()
```

```python
df_languages= df[Languages_used_in_survey]
df_languages.head()
```

```python
df2 = df[["Age", "ConvertedComp"]].dropna()
cond = (df["Age"] >= 20 )  & (df["Age"] <= 60 )
df2 = df2.loc[cond, :]
df3 = df2.groupby("Age").mean()
```

```python
# df3.head().to_frame()
cond2 = df3["ConvertedComp"] < 1069000.0
df3 = df3.loc[cond2,:]
df3.head()
df3
```

```python
sal=df3["ConvertedComp"].reset_index()
```

```python
sal = sal.assign(expenses = lambda x:x.ConvertedComp - (x.Age*1000))
sal
```

```python
ax = sal.plot(x="Age", y="ConvertedComp", linewidth=0.4, figsize=(10,6))
sal.plot(x="Age", y="expenses", linewidth=0.4, ax=ax)
plt.fill_between(sal["Age"],sal["ConvertedComp"], sal["expenses"],
                 where=((sal["ConvertedComp"]>sal["expenses"]) &
                        (sal["ConvertedComp"]<200000) &
                        (sal["ConvertedComp"]>50000)),
                 interpolate=True, alpha=0.25,
                 label="Gap")
plt.legend()
plt.show()
```
