<font color = sky_blue>*iloc[index]* </font> retrieves the index of the column mentioned in prefix

### To retrive a M x N dataframe
>dataframe.iloc[[rows], [column]]</br>
>`iloc` is used to retrive rowns and columns by indices


```python
reviews.iloc[[5, 6, 7], [1,2,3]]  # retrieve 5,6,7th row and 1,2,3rd column
```

### Retrive by location
>`loc` retrives rows and columns based on key

```python
new_df = reviews.loc[[1,20,47,76], ["country", "points", "price"]]
new_df
```

### Filter certain column and return the values
>Suppose we need to search the column country for 'India' entries

```python
Indian = reviews.loc[reviews.country == "India"]
Indian
```

### To add additional filter operation use ampersand &
> Below we used 2 conditions the wines should be from country India and price should be 12

```python
Indian = reviews.loc[(reviews.country == "India") & (reviews.price == 12)]
Indian
```
