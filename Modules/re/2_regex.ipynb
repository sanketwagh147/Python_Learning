```python
# Sample code to generate random sentences 
import random
import string
num_words = 8
string_list = []
for i in range(10):
    word_1 = ''.join(random.choices(string.ascii_lowercase + string.digits, k=num_words))  # k is keyword parameter which states length of word and is mandatory to define it as k = <any int>
    word_2 = ''.join(random.choices(string.ascii_lowercase + string.digits, k=num_words))
    word_3 = ''.join(random.choices(string.ascii_lowercase , k=num_words))
    sentence = word_1 +" " + word_2 + " " + word_3
    string_list.append(sentence)
string_list
```

```python
# To find a pattern which has words in range
random_sentence_list = ['aqlm tdet cahm',
 'uqps ywuo yrwe',
 'xrrf dman qhdz',
 'rtrg cgkz looe',
 'qqgk vvpl zrnk',
 'xlaf cfcp kpkf',
 'nqgh qznm zisw',
 'lvni omnf lgwd',
 'dipc aiie vbxf',
 'mfec ffvc knzs']
pattern_set3 = "[u-x]" # Hyphen(-) is used to set a range this means letters from u to x
result = []
for sentence in random_sentence_list:
    result.append(re.search(pattern_set3, sentence))
for i,res in enumerate(result, start=1):  # start = 1 starts the index with 1 instead of 0
    print(i,":",res)
```

## Not in Regex set `"[^ ]"`(all charcter except the ones followed by `^`)
>Caret inside a square bracket or Regex set simply means not 

```python
random_sentence_list = ['abccabccabcc',
 'uqps ywuo yrwe',
 'xrrf dman qhdz',
 'rtrg cgkz looe',
 'qqgk vvpl zrnk',
 'caba baca abac',
 'nqgh qznm zisw',
 'lvni omnf lgwd',
 'dipc aiie vbxf',
 'mfec ffvc knzs']
pattern_set4 = "[^a-c]" # ^ is used inside brackets search for any characters apart from mentioned and match if any
result = []
for sentence in random_sentence_list:
    result.append(re.search(pattern_set4, sentence))
for i,res in enumerate(result, start=1):  # start = 1 starts the index with 1 instead of 0
    print(i,":",res)
```

### notice sentence 1 do not have any characters apart from a-c thus the result is None
*<font color= red>Important Note:</font> Notice sentence 6 do not contain any letters apart from a-c but it includes a space thus the result is " " as this pattern searches for any character except the ones mentioned*

## Special Sequences backslash:`\`
>`\d` : only digits [0-9] </br>
>`\w` only words characters [0-9, a-Z, _ ]


## `\d`

```python
pattern_5 ="\d"
string_3 = "123 testing"
string_4 = "0 is first"
string_5 = "what"
(re.search(pattern_5, string_3),
 re.search(pattern_5, string_4),
 re.search(pattern_5, string_5))
```

```python
#### string 5 returns None as their is no digit(0-9) in the string
```

```python
random_sentence_list = ['uczabcyi ubnazaqb odsyjrrg',
 'dkoypjjn cne8wqq7 vkcbtiev',
 'iz82nb81 3y3z1fpl odyzaooc',
 '0zdfrugw s2ywn9jg vshbeghg',
 'y28welve zgh2sxlg mxhbnnfu',
 'abaccxab jyciqaqf whbmcmto',
 '5ud25dpm ddrq3jzf jcctanlc',
 'dg3h9ezn yef5m7fq gfxxhjds',
 'b9ee2pbg zrbnvj7s abildbbm',
 'q4tzgxpf 328m46cb kuxhqysc']
pattern_5 = "\d" # `\d` is used to find for digits 0-9
result = []
for sentence in random_sentence_list:
    result.append(re.search(pattern_5, sentence))
for i,res in enumerate(result, start=1):  # start = 1 starts the index with 1 instead of 0
    print(i,":",res)
```

#### This finds the string of any digits in sentence
Sentence 6 and 1 do not have a single digit hence return None

## `\w` search word (a-Z) , digits(0-9) and underscore(_)

```python
string_6 = "this_is all I got!"
string_7 = "^**$%"
pattern_6 = "\w"
re.findall(pattern_6,string_6)  # return all above mentioned words
```

```python
re.findall(pattern_6, string_7)  # as the word doesn not contain single character it returns empty list
```

## The difference between search and findall is:
>search returns a match object of first occurence including span</br>
>findall returns a list of characters matched
