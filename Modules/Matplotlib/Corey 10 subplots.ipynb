```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('datasub.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (axis1,axis2) = plt.subplots(nrows=2, ncols=1, sharex=True)   # by default 1 or however we like

axis1.plot(ages, py_salaries, label='Python')
axis1.plot(ages, js_salaries, label='JavaScript')
axis2.plot(ages, py_salaries, label='Python')
axis2.plot(ages, js_salaries, label='JavaScript')
axis1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

axis2.set_xlabel('count')
axis2.set_ylabel('boo')



axis1.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
