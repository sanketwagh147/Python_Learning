```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(filepath)
```

## Density Plot

```python
ax= df[["reading score", "math score"]].plot(kind="density", figsize=(12,6))
# Other addtions
# ax.axvline(df['writing score'].mean(), color="red", label="mean")
# ax.axvline(df['writing score'].median(), color="green", label="median")
# ax.axvline(df['writing score'].max(), color="green", label="max")
# ax.axvline(df['writing score'].min(), color="blue", label="mean")
plt.xlim((0,125))
ax.set_facecolor("white")
plt.legend()
plt.rcParams['axes.facecolor'] = 'white'
plt.savefig("denplot.png")
```

## Histogram

```python
ax = df["writing score"].plot(kind='hist', figsize=(14,6), color="#FBFBF2", ec="red")
ax.set_ylabel('Marks')
ax.set_xlabel('Num_students')
plt.savefig("denplot.png")
# plt.show()
```
